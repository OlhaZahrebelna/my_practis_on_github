```python
import datetime
```

```python
def create_datetime(row):
    if len(row) == 19:
        try:
            return datetime.datetime.strptime(row, '%Y-%m-%d %H:%M:%S')
        except ValueError:
            return datetime.datetime.strptime(row, '%m.%d.%Y %H:%M:%S')
    elif len(row) == 16:
        try:
            return datetime.datetime.strptime(row, '%d.%m.%Y %H:%M')
        except ValueError:
            return datetime.datetime.strptime(row, '%m.%d.%Y %H:%M')
```

```python
df['Timestamp'] = df['Timestamp'].apply(create_datetime)
```

14. Як зростає зарплата з роками? Для цього нам треба відформатувати колонку "Загальний стаж роботи за спеціальністю" аналогічно до того, як ми це робили для колонки 'Знання англійської мови' і потім вивести медіанні значення для кількості років стажу.

```python
def convert_to_number(value):
    if 'Пів' in value:
        return 0.5
    elif 'менше як 3 місяці'.lower() in value.lower():
        return 0.25
    elif 'місяці' in value.lower():
        return 0.25
    elif 'рік' in value.lower() or 'роки' in value.lower():
        return float(value.split()[0])
    elif 'більше' in value.lower():
        return 15
    elif value == 'nan':
        return np.nan
    else:
        return value

ordered_experience = ['Менше як 3 місяці', '3 місяці', 'Пів року', '1 рік', '2 роки',
                      '3 роки', '4 роки', '5 років', '6 років', '7 років', '8 років',
                      '9 років', '10 років', '11 років', '12 років', '13 років',
                      '14 років', '15 і більше років']

# Приводимо колонку "Загальний стаж роботи за спеціальністю" до категоріального типу даних
df['Загальний стаж роботи за спеціальністю'] = pd.Categorical(df['Загальний стаж роботи за спеціальністю'],
                                                              categories = ordered_experience, ordered=True)

# Створюємо нову колонку з числовими значеннями стажу для розрахунків
df['Стаж у роках'] = df['Загальний стаж роботи за спеціальністю'].apply(convert_to_number)

# Розраховуємо медіанні зарплати для кожної категорії стажу
median_salary_by_experience = df.groupby('Загальний стаж роботи за спеціальністю')['Зарплата у $$$ за місяць, лише ставка після сплати податків'].median()

median_salary_by_experience
```
