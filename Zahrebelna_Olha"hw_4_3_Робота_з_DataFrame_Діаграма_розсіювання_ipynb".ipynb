Ми маємо графік розсіювання з позитивною кореляцією. Це означає, що зі збільшенням температури фактичної також зростає температура, що сприймається.

У діапазоні від 0 до 16 градусів за віссю x і від 0 до 19 градусів за віссю y графік показує певне розсіювання: зазвичай температура відчувається такою ж, як і є, або теплішою. Тут ми спостерігаємо різні відчуття однієї й тієї ж температури. У діапазоні від 16 до 24 градусів за віссю x температура, що відчувається, є більш стабільною — приблизно 18 градусів, незалежно від фактичної температури.

Однак, у діапазоні від 24 до 36 градусів спостерігаються аномалії: температура, що відчувається, іноді знижується до приблизно 13 градусів. Це може бути пов'язано з певними умовами, наприклад, наявністю позначки 4 у колонці "weather". У проміжку від 24 до 42 градусів ситуація схожа на ту, що спостерігається в діапазоні від 0 до 16 градусів, де також спостерігається розсіювання даних.

```python
df_atemp_anomaly = df[(df['atemp'] >= 10) & (df['atemp'] <= 13) & (df['temp'] >= 24) & (df['temp'] <= 36)]
df_atemp_anomaly.head(24)
```
